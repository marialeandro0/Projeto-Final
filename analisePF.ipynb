```python
df_sorted = df.sort_values(by='Female to male ratio of time devoted to unpaid care work (OECD (2014))', ascending=False)
```

```python
top5_maior_proporcao = df_sorted.head(5)
top5_menor_proporcao = df_sorted.tail(5)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(top5_maior_proporcao['Entity'], top5_maior_proporcao['Female to male ratio of time devoted to unpaid care work (OECD (2014))'], color='skyblue')
plt.title('Top 5 Países com Maior Proporção de Tempo Dedicado a Trabalho de Cuidado Não Remunerado por Mulheres')
plt.xlabel('Países')
plt.ylabel('Proporção')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(top5_menor_proporcao['Entity'], top5_menor_proporcao['Female to male ratio of time devoted to unpaid care work (OECD (2014))'], color='salmon')
plt.title('Top 5 Países com Menor Proporção de Tempo Dedicado a Trabalho Não Remunerado por Mulheres')
plt.xlabel('Países')
plt.ylabel('Proporção')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
paises_sulamericanos_lista = [
    'Argentina', 'Bolivia', 'Brazil', 'Chile', 'Colombia', 'Ecuador',
    'Guyana', 'Paraguay', 'Peru', 'Suriname', 'Uruguay', 'Venezuela'
]
```

```python
paises_sulamericanos = df[df['Entity'].isin(paises_sulamericanos_lista)]
```

```python
plt.figure(figsize=(12, 8))

for pais in paises_sulamericanos_lista:
    proporcao = df[df['Entity'] == pais]['Female to male ratio of time devoted to unpaid care work (OECD (2014))']
    plt.bar(pais, proporcao, label=pais)

plt.title('Proporção de Tempo Dedicado a Trabalho Não Remunerado (América do Sul)')
plt.xlabel('Países')
plt.ylabel('Proporção')
plt.legend(title='Países', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
